```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('diabetes.csv')
df.shape
```

```python
sns.kdeplot(data=df,x='Pregnancies')
plt.show()
```

```python
sns.kdeplot(data=df,x='Pregnancies',hue='Outcome')
plt.show()
```

```python
sns.kdeplot(data=df,x='Pregnancies',hue='Outcome',multiple='stack')
plt.show()
```

```python
sns.kdeplot(data=df,x='Pregnancies',multiple='stack',alpha=0.2)
plt.show()
```

```python
sns.kdeplot(data=df,y='Pregnancies',multiple='stack',alpha=0.2)
sns.boxplot(x='Pregnancies',data=df,showmeans=True)
plt.show()
```
