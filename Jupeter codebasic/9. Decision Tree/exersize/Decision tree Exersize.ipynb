```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
new_df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
new_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
new_df['sex_n'] = le_sex.fit_transform(new_df['Sex'])
new_df.head()
```

```python
new_df = new_df.drop(['Sex'],axis='columns')
```

```python
new_df.Age = new_df.Age.fillna(new_df.Age.mean())
```

```python
new_df
```

```python
inputs = new_df.drop(['Survived'],axis='columns')
inputs
```

```python
target = new_df.Survived
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,target,train_size=0.7)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
