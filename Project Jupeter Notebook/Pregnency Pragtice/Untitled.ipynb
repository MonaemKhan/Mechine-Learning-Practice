```python
import pandas as pd
df = pd.read_csv('Pregnancy Risk.csv')
df.sample(5)
```

```python
df1 = df.drop(['Unnamed: 0'],axis=1)
df1.head(5)
```

```python
new = pd.read_csv('Para_Gravida.csv')
new.shape
```

```python
length = df1.shape[0]
length
```

```python
df1['Delivery Type'][0]
```

```python
import random
random.randint(0,3)
```

```python
for i in range(length):
    n = random.randint(1,2)
    if df1['Delivery Type'][i] == 'No_delivary':
        new.Para[i] = 0
        new.Gravida[i] = 1
    else:
        if df1.Age[i]<20:
            if df1.Risk[i] == 'Yes':
                new.Para[i] = 1
                new.Gravida[i] = 2
        else:
            new.Para[i] = n
            new.Gravida[i] = n+1
```

```python
new = new.astype(int)
```

```python
n = pd.concat([df1,new], axis=1, join="inner")
n
```
