```python
x = ['SVM','GNB','MNB','DT','RF']
y_train = [71,75,76,100,100]
y_test = [67,71,72,86,91]
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(121, projection='3d')

ax.bar(x,y_train,zs = [1,1,1,1,1],zdir='y')
```

```python
name = x
price = y_train

# Figure Size
fig, ax = plt.subplots(figsize =(10, 5))

# Horizontal Bar Plot
ax.barh(name, price)

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(b = True, color ='gray',linestyle ='-.', linewidth = 0.5,alpha = 0.2)

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:plt.text(i.get_width()+0.2, i.get_y()+0.5,str(round((i.get_width()), 2))+'%',
                             fontsize = 14, fontweight ='light',color ='black')

# Add Plot Title
ax.set_title('Comparative Results For Training Accuracy',loc ='left', )

# Add Text watermark
fig.text(0.9, 0.15, '', fontsize = 14,color ='black', ha ='right', va ='bottom',alpha = 0.7)
plt.savefig("Algorithm Compairson on Training Data.png")
# Show Plot
plt.show()
```
