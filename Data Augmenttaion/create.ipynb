```python
import pandas as pd
```

```python
df = pd.read_csv('Pregnancy Risk.csv')
df.head()
```

```python
df.shape
```

```python
df1 =  df[df.Risk == 'Yes']
df1.shape
```

```python
print(df1.shape)
df1.head()
```

```python
df1_age = df1.Age.reset_index(drop=True)
print(df1_age.shape)
```

```python
df1_height = df1.Height.reset_index(drop=True)
df1_height.shape
```

```python
bmi = df1.BMI.reset_index(drop=True)
print(bmi.shape)
```

```python
df1_weight = df1.Weight.reset_index(drop=True)
print(df1_weight.shape)
```

```python
df1_blood = df1['Blood Group ( is Negative?)'].tolist()
df1_blood = pd.DataFrame(df1_blood)
df1_blood.rename(columns={0:'Blood Group ( is Negative?)'}, inplace=True)
df1_blood.reset_index(drop=True)
```

```python
df1_del = df1['Delivery Type'].tolist()
df1_del = pd.DataFrame(df1_del)
df1_del.rename(columns={0:'Delivery Type'}, inplace=True)
df1_del.reset_index(drop=True)
```

```python
part1 = pd.concat([df1_age,df1_height,df1_weight,bmi,df1_blood,df1_del],axis=1)
```

```python
part1
```

```python
df1.head(0)
```

```python
df1_up = df1['Systolic Blood Pressure']
print(df1_up.shape)
df1_up = df1_up.sample(frac=1).reset_index(drop=True)
df1_up
```

```python
df1_dn = df1['Diastolic Blood Pressure']
print(df1_dn.shape)
df1_dn = df1_dn.sample(frac=1).reset_index(drop=True)
df1_dn
```

```python
part2 = pd.concat([part1,df1_up,df1_dn],axis=1)
part2
```

```python
part3 = df1.drop(['Age','Height','Weight','BMI','Blood Group ( is Negative?)','Delivery Type','Systolic Blood Pressure','Diastolic Blood Pressure'],axis=1).reset_index(drop=True)
print(part3.shape)
part3
```

```python
final_part = pd.concat([part2,part3],axis=1)
```

```python
final_part
```

```python
final_part = final_part.sample(frac=1).reset_index(drop=True)
final_part
```

```python
final_part.to_csv('Pregnancy_risk.csv')
```
