```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('diabetes.csv')
df.shape
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df['Outcome'].value_counts()
```

```python
df[df['Pregnancies']>10].shape
```

```python
range = len(df.columns)-1
for col in df.columns[:range]:
    plt.title(col)
    sns.boxplot(
        x=col,
        data=df,
        showmeans=True
    )
    plt.savefig('./With outliners/'+col+'.png')
    plt.show()
```
