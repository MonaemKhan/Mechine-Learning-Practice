```python
import seaborn as sn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
range = len(df.columns)-1
for col in df.columns[:range]:
    plt.title(col)
    sn.boxplot(
        x=col,
        data=df,
        showmeans=True
    )
    plt.show()
```

```python
num_list = ['Age','SystolicBP','DiastolicBP','BS','BodyTemp','HeartRate']
fig = plt.figure(figsize=(10,10))
j=0;
for i in num_list:
    plt.subplot(3,3,j+1)
    plt.title(i)
    sn.distplot(df[df['RiskLevel'] == 'low risk'][i],hist=False,kde=True,kde_kws={'multiple':'stack','alpha':0.2}) 
    sn.distplot(df[df['RiskLevel'] == 'mid risk'][i],hist=False,kde=True,kde_kws={'multiple':'stack','alpha':0.2})
    sn.distplot(df[df['RiskLevel'] == 'high risk'][i],hist=False,kde=True,kde_kws={'multiple':'stack','alpha':0.2})
    j = j+1
plt.tight_layout()
plt.show()
```
